#tree
디렉토리 구조를 트리 형태로 시각적으로 보여주는 명령어야. 파일과 폴더의 계층 구조를 한눈에 보기 좋게 출력해줘서 프로젝트 구조 파악할 때 유용

tree (현재 디렉토리를 기준으로 모든 하위 폴더와 파일을 트리 형태로 출력)

# tree 예시

tree /path/to/directory (특정 디렉토리 보기)

tree -L 2 (깊이 제한, 설명 : 2단계까지만 출력)

tree -f (파일만 보기, 파일 경로까지 포함해서 출력)

tree -d (파일만 보기, 디렉토리만 출력)

tree -a (숨김 파일 포함)

tree -h (용량 표시, 파일 크기를 사람이 읽기 쉽게 표시)

tree -I "node_modules" (특정 패턴 제외, node_modules 제외)

tree > structure.txt (결과를 파일로 저장,)


sudo apt install tree (설치X)(Ubuntu / Debian)


brew install tree (설치X) (macOS (Homebrew))
예시 출력

├── src

│   ├── main.js

│   └── utils.js

├── package.json

#Ctrl+c 실행 취소

# ls (중요O)
Windows의 dir과 같은 역할로,해당 디렉터리에 있는 파일의 목록을 나열

예) # ls /etc/systemd

# cd (중요O)
디렉터리를 이동

예) # cd ../etc/systemd

#pwd (중요O)
현재 디렉터리의 전체 경로를 출력

#touch (중요O)
크기가 0인 새 파일을 생성, 이미 존재하는 경우 수정 시간을 변경

예) # touch abc.txt

#rm (중요O)
파일이나 디렉터리를 삭제

예) # rm -rf abc

#cp (중요O)
파일이나 디렉터리를 복사

예) # cp abc.txt cba.txt

#mv (중요O)
파일과 디렉터리의 이름을 변경하거나 위치 이동 시 사용

예) mv abc.txt www.

#mkdir (중요O)
새로운 디렉터리를 생성

예) # mkdir abc

#rmdir
디렉터리를 삭제. (단, 비어 있어야 함)

예) # rmdir abc

#cat (중요O)
텍스트로 작성된 파일을 화면에 출력

예) # cat a.txt b.txt

#head, tail
텍스트로 작성된 파일의 앞 10행 또는 마지막 10행만 출력

예) # head /etc/systemd/user.conf

#more
텍스트로 작성된 파일을 화면에 페이지 단위로 출력

예) # more /etc/systemd/system.conf

#less
more와 용도가 비슷하지만 기능이 더 확장된 명령

예) # less /etc/systemd/system.conf

#file (중요O)
File이 어떤 종류의 파일인지를 표시

예) # file /etc/systemd/system.conf

#clear
터미널 화면을 깨끗하게 지워줌

예) # clear
